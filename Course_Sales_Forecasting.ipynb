```python
def smape(ActVal, ForeVal):
    smape_score = 100/len(ForeVal) * (np.sum(np.abs(ForeVal-ActVal)/(np.abs(ActVal)+np.abs(ForeVal))/2))
    return smape_score
```

```python
from sklearn.model_selection import TimeSeriesSplit
def evaluate(model):
    tscv = TimeSeriesSplit(n_splits = 3, test_size = test_size)
    scores = []
    for i, (train_ind, test_ind) in enumerate(tscv.split(df_train)):
        X_train, y_train = X.iloc[train_ind], y.iloc[train_ind]
        X_test, y_test = X.iloc[test_ind], y.iloc[test_ind]
        
        model.fit(X_train, y_train)
        pred = model.predict(X_test)
        
        smape_score = smape(y_test, pred)
        scores.append(smape_score)
        
        print('{0} sMAPE: {1: .3f}'.format(model.__class__.__name__, np.mean(scores)))
        return np.mean(scores)
```

```python
def best_params(obj_func, space):
    trials = Trials()
    
    best_par = fmin(fn = obj_func, space = space,
               algo = tpe.suggest, max_evals = 50,
               trials = trials, rstate = np.random.default_rng(seed = 30))
    print("Best Params:", best_par)
    return best_par
    
```

```python
def weighted_ensemble():
    final_preds = []
    for i in range(3):
        denom = (1/lgbmScore[i]) + (1/grboostScore[i]) + (1/catboostScore[i]) + (1/xgbScore[i])
        weight_lgbm = (1/lgbmScore[i])/denom
        weight_gr = (1/grboostScore[i])/denom
        weight_cb = (1/catboostScore[i])/denom
        weight_xgb = (1/xgbScore[i])/denom
        lgbm_pred_i = np.multiply(weight_lgbm, lgbm_pred[i])
        grboost_pred_i = np.multiply(weight_gr, grboost_pred[i])
        catboost_pred_i = np.multiply(weight_cb, catboost_pred[i])
        xgb_pred_i = np.multiply(weight_xgb, xgb_pred[i])
        pred = [ a+b+c+d for a,b,c,d in zip(lgbm_pred_i, grboost_pred_i, catboost_pred_i, xgb_pred_i)]
        final_preds.append(pred)
    final_pred = [sum(x) for x in zip(*final_preds)]
    return np.divide(final_pred, 3)
```

```python
final_preds = weighted_ensemble()
print('Mean of Weighted Ensembled Prediction: {0: .3f}'.format(final_preds.mean()))
```
